```python
import datetime
data=pd.read_table(r"https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data",sep = "\s+",parse_dates = [[0,1,2]])
data.head()
```

```python
#step-4
def yearError(x):
  year = x.year - 100 if x.year > 1989 else x.year
  return datetime.date(year, x.month, x.day)
data['Yr_Mo_Dy'] = data['Yr_Mo_Dy'].apply(yearError)
data.head()
```

```python
#step 5
data["Yr_Mo_Dy"] = pd.to_datetime(data["Yr_Mo_Dy"])
data = data.set_index('Yr_Mo_Dy')
data.head()
```

```python
#step6
data.isnull().sum()
```

```python
data.notnull().sum()
```

```python
#step8
data.sum().sum() / data.notna().sum().sum()
```

```python
loc_stats=data.describe(percentiles=[])
loc_stats
```

```python
#step 10
day_stats = pd.DataFrame()
day_stats['min'] = data.min(axis = 1)
day_stats['max'] = data.max(axis = 1)  
day_stats['mean'] = data.mean(axis = 1)
day_stats['std'] = data.std(axis = 1) 

day_stats.head()
```

```python
#step11
data.mean()
```

```python
data.groupby(data.index.to_period('A')).mean()
```

```python
data.groupby(data.index.to_period('W')).mean()
```

```python
week_wise = data.resample('W').agg(['min','max','mean','std'])

week_wise.loc[week_wise.index[1:53], "RPT":"MAL"] .head(10)
```
